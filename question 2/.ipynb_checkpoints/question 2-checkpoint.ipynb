## function define

```python

from numpy import sinc
x1c  = lambda t: sinc(t/6)
x2c = lambda t: (sinc(t/12))**2
x3c = lambda t: np.cos(np.pi*t/12)
signals = [x1c, x2c, x3c]
```

```python
band_times = 10
resolution = 200
tstart = -band_times*np.pi
tstop = -tstart

t = np.linspace(tstart, tstop, resolution)
fig,ax = plt.subplots(4, 1)
ax[0].plot(t, sinc(t), label="sinc(t)")
ax[1].plot(t, x1c(t), label="x1c(t)")
ax[2].plot(t, x2c(t), label="x2c(t)")
ax[3].plot(t, x3c(t), label="x3c(t)")
for axis in ax:
    axis.legend()
fig.suptitle('Signals in time', fontsize=16)
```

```python
delta_epsilon = 0.5
Xj1 = lambda s: 6*(abs(s) <= np.pi/6) 
# Xj2 = lambda s: (s*72*pi/12 + 12) if (-pi/6 <= s and s<= 0) else ((-1*s*72*pi/12 + 12) if (0 <= s and s<= pi/6) else 0 )
def Xj2_helper(s):
    if (-pi/6 <= s and s<= 0):
        return s*72*pi/12 + 12
    elif (0 <= s and s<= pi/6):
        return -1*s*72*pi/12 + 12  
    else:
        return 0
Xj2 = lambda s: np.array([Xj2_helper(i) for i in s])
    

delta = lambda s: 1*(abs(s) <= delta_epsilon)
Xj3 = lambda s:np.pi*(delta(s-np.pi/12)+delta(s+np.pi/12))
Xj = [Xj1, Xj2, Xj3]
```
