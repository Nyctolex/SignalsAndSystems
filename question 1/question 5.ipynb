```python
roar = lambda x, x_min, x_max:  (x+x_min)%(x_max-x_min)+x_min
s = lambda t: -1+np.sqrt(-t)*np.exp(t) if (-1 <= t and t<=0) else 1-np.sqrt(t)*np.exp(-t)
x = lambda t: np.array(list(map(s, roar(t, -1, 1))))
t = np.linspace(-4, 4, 1000)
fig, ax = plt.subplots(1, 1)
ax.plot(t, x(t), label="x(t)")
ax.set_xlabel("t")
ax.set_ylabel("x(t)")

ax.legend()
title = 'Question 1 section a'
ax.set_title(title)
plt.tight_layout()
plt.savefig(name(title))
```

```python
shift_f = lambda f, T, N, n: f(-T/2+((2*n-1)*T)/(2*N))
approax_int = lambda f, T, N:(T/N)*sum(shift_f(f, T, N, np.array(range(1, N))))

def ak(k,f,T,N):
    t = sp.Symbol('t', real=True)
    cos_exp = lambdify(t,sp.cos(2*np.pi*k*t/T))
    integral_exprestion = lambda t: f(t)*cos_exp(t)
    a = approax_int(integral_exprestion, T, N)
    if k==0:
        a = a/T
    else:
        a = (2/T)*a
    return a

def bk(k,f,T,N):
    t = sp.Symbol('t',  real=True)
    sin_exp = lambdify(t,sp.sin(2*np.pi*k*t/T))
    integral_exprestion = lambda t: f(t)*sin_exp(t)
    return (2/T)*approax_int(integral_exprestion, T, N)

def get_a_series(f, T, N, last_K):
    a = []
    K = np.array(range(0, last_K))
    for i in K:
        a.append(ak(i, f, T, N))
    a = np.array(a)
    return a

def get_b_series(f, T, N, last_K):
    b = [0]
    K = np.array(range(1, last_K))
    for i in K:
        b.append(bk(i, f, T, N))
    b = np.array(b)
    return b

def get_xm_function(M, a_series, b_series,T):
    symbolic_t = sp.Symbol('t', real=True)
    K = np.array(list(range(0, M)))
    w0 = (2*np.pi)/T
    cos_sum = 0
    xm = 0
    for k in K:
        xm +=a_series[k]*sp.cos(w0*k*symbolic_t)+b_series[k]*sp.sin(w0*k*symbolic_t)
    return lambdify(symbolic_t, xm)

# N = 200
# M= 5
# T = 2*np.pi
# f = lambda x: np.sin(x)
# ks = np.array(list(range(0, M)))
# a = get_a_series(f, T, N, M)
# b = get_b_series(f, T, N, M)
# xm = get_xm_function(M, a, b,T)

# t = np.linspace(0, T, 100)
# fig, (series, ax) = plt.subplots(2, 1)
# series.bar(ks,a, label='ak')
# series.bar(ks,b, label='bk')
# series.legend()
# ax.plot(t, f(t), label='f')
# ax.plot(t, xm(t), label='x')
# ax.legend()
```
